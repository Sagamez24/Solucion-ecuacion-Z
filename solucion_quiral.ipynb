```python
import numpy as np


############################################################
 
  
def valor_m(n):
    """
    valor_m muestra el la cantidad de L y K que se produciran 
    """
    if n%2==0:
        m = n/2 -1
    else:
        m = (n-3)/2
    return m

def sol_quiral(n):
    """
    Esta funcion devuelve los vetoreslikes v_mas y v_menos o u_mas y u_menos
    a partir de los arreglos L y K tomados de la lista aleatoria con valores
    entre -3 y 3
    """
    m = int(valor_m(n))
    lista = np.random.randint(low=-3, high=4, size=30).tolist() 
    print(lista)
#    print(lista)

    if m%2 == 0:
        L = lista[:m]
        K = lista[:m]
                
        v_mas = np.append(L[0],K)
        v_menos = np.append([0,0],L)

        v_mas = np.append(v_mas, -np.append(L[0],K))
        v_menos = np.append(v_menos, -np.array(L))

        Z_sol = np.append([v_mas], [v_menos], axis=0)

    else:
        L = lista[:m]
        K = lista[:m+1]
        
        u_mas = np.append([0], K)
        u_menos = np.append(L, [K[0],0])
        
        u_mas = np.append(u_mas, -np.array(K))
        u_menos = np.append(u_menos, -np.append(L, K[0]))

        Z_sol = np.append([u_mas], [u_menos], axis=0)
        
    return Z_sol

def operacion(z_sol):
    """
    función vector like
    
    Return
    ----------
    MCD : retorna el maximo comun divisor 
        DESCRIPTION.
    resultado : retorna la operación
    
    """
    op = np.sum(Z_sol[0] * z_sol[1]**2) * z_sol[0] - np.sum(z_sol[0]**2 * z_sol[1]) * z_sol[1]
    op = sorted(op)  ##Ordenando la solución (tanto quiral como "vectorlike")
    gcd = np.gcd.reduce(op)  ##Máximo común divisor de la solución
    
    if np.all((op == 0)): ##Para evitar un arreglo de nan
        resultado = op
    else:
        resultado = op/gcd
    return (gcd, resultado)

def sistema_ecu(z):
    """
    esta funcion verifica que se cumpla que z^3+..+zn^3=0 y que z+...+zn=0
    
    Parameters
    ----------
    x : Es la solucion correspondiente a Z
    MCD : es el maximo comundivisor del conjunto de valores de z

    Returns
    -------
    x : Es la solucion correspondiente a Z
    MCD : es el maximo comundivisor del conjunto de valores de z
    """

    ecu_1 = np.sum(z[1]**3)
    ecu_2 = np.sum(z[1])
    
    if ecu_1 == 0 and ecu_2 == 0:
       return(z[1])

val = 30
def chequeo_z(x, val):
        for s in x:
            if abs(s) >= val:
                return()
        return(x)       

n = int(input('incerte el valor para n:'))
m = int(valor_m(n))
Z_sol = sol_quiral(n)
z = operacion(Z_sol)
x = sistema_ecu(z)
#x = x[np.logical_and(x<30,x>-30)]
x = chequeo_z(x, val)
print(f'el valor para m es:{m}')
print(f'El arreglo vectorlike es:{Z_sol}')
print(f'la solucion z es:{z}')
print(f'la condicion se culple y el conjunto quiral es z: {x}')
print(f'el valor de y es:{x}')
```
